```python
import math

def calculate_soh(ir, qc, qd, tavg, tmin, cycle_no, cycle_life, charge_time):
    # Calculate the Peukert exponent (typically around 1.1 to 1.3)
    peukert_exponent = 1.2

    # Calculate the discharge rate
    discharge_rate = qd / (charge_time * 3600)  # Convert charge_time to seconds

    # Calculate the discharge capacity using Peukert's equation
    peukert_capacity = qc * (cycle_life / cycle_no) * ((discharge_rate ** peukert_exponent) / (peukert_exponent * ir))

    # Calculate the actual capacity degradation
    actual_capacity = peukert_capacity * (tavg / tmin)

    # Calculate SOH as a percentage
    soh_percentage = (actual_capacity / qc) * 100

    return soh_percentage

# Example parameters
IR = 0.016742354 # Internal resistance (Ohm)
QC = 1.07   # Initial capacity (Ah)
QD = 1.07  # Discharge capacity (Ah)
Tavg = 31.87 # Average temperature (C)
Tmin = 29.56  # Minimum temperature (C)
cycle_no = 1  # Current cycle number
cycle_life = 1852  # Expected cycle life
charge_time = 13.34  # Charge time (hours)

# Calculate SOH
soh = calculate_soh(IR, QC, QD, Tavg, Tmin, cycle_no, cycle_life, charge_time)
print(f"SOH: {soh:.2f}%")
```
